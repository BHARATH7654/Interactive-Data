# Final Project: Student Performance Analysis

## Overview
This project aims to explore the impact of demographic factors, parental education level, and test preparation courses on student performance in math, reading, and writing.

### Research Questions
1. **Influence of Demographic Factors:** How do demographic factors like gender, race/ethnicity, and socioeconomic status (based on lunch program participation) influence student performance in math, reading, and writing?
2. **Impact of Parental Education Level:** What is the impact of parental education level on student academic achievement?
3. **Effect of Test Preparation Courses:** Does participation in test preparation courses significantly improve student performance?

### Summary of Findings
1. **Influence of Demographic Factors:**
   - Gender: On average, female students perform better in reading and writing compared to male students, while math scores are relatively similar.
   - Race/Ethnicity: There is variation in performance across racial/ethnic groups, with some groups performing significantly higher in certain subjects.
   - Socioeconomic Status: Students who qualify for free/reduced lunch tend to score lower in all subjects compared to those who do not.

2. **Impact of Parental Education Level:**

3. **Effect of Test Preparation Courses:**

## Student Performance Analysis

### Writing, Reading, and Math Pass

- **Writing Pass**: Checks if a student's writing score is less than 40.
- **Reading Pass**: Checks if a student's reading score is less than 40.
- **Math Pass**: Checks if a student's math score is less than 40.

### Grades

The code calculates the total marks and percentage for each student and assigns a grade based on the following criteria:
- 'A' for 80% and above
- 'B' for 70% to 79%
- 'C' for 60% to 69%
- 'D' for 50% to 59%
- 'E' for 40% to 49%
- 'F' for below 40% or if the student has failed in any subject.
The results provide insights into the overall performance of students and help identify areas where additional support might be needed.

```python
# Define pass/fail columns
df['Write_Pass'] = np.where(df['writing score'] < 40, 'F', 'P')
df['Read_Pass'] = np.where(df['reading score'] < 40, 'F','P')
df['Math_Pass'] = np.where(df['math score'] < 40, 'F', 'P')

# Print counts for each pass category
print('Writing Pass/Fail Counts:\n', df['Write_Pass'].value_counts())
print('Reading Pass/Fail Counts:\n', df['Read_Pass'].value_counts())
print('Math Pass/Fail Counts:\n', df['Math_Pass'].value_counts())

# Create a figure with three subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot Writing Pass/Fail
sns.countplot(x='parental level of education', data=df, hue='Write_Pass', ax=axs[0], palette='bright')
axs[0].set_title('Writing Pass/Fail by Parental Education')
axs[0].set_xlabel('Parental Level of Education')
axs[0].set_ylabel('Count')
axs[0].tick_params(axis='x', rotation=45)

# Plot Reading Pass/Fail
sns.countplot(x='parental level of education', data=df, hue='Read_Pass', ax=axs[1], palette='bright')
axs[1].set_title('Reading Pass/Fail by Parental Education')
axs[1].set_xlabel('Parental Level of Education')
axs[1].set_ylabel('Count')
axs[1].tick_params(axis='x', rotation=45)

# Plot Math Pass/Fail
sns.countplot(x='parental level of education', data=df, hue='Math_Pass', ax=axs[2], palette='bright')
axs[2].set_title('Math Pass/Fail by Parental Education')
axs[2].set_xlabel('Parental Level of Education')
axs[2].set_ylabel('Count')
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()


df['All3_Pass'] = df.apply(lambda x : 'Fail' if x['Math_Pass'] == 'F' or x['Read_Pass'] == 'F' or x['Write_Pass'] == 'F' else 'Pass', axis=1)
print(' counts \n', df['All3_Pass'].value_counts())
m = sns.countplot(x='parental level of education', data=df, hue='All3_Pass')
# Percentage of Marks
df['Total_Marks'] = df['math score'] + df['reading score'] + df['writing score']
df['Percentage'] = df['Total_Marks'] / 3

def GetGrade(Percentage, All3_Pass):
    if (All3_Pass == 'Fail'):
        return 'F'
    if (Percentage >= 80):
        return 'A'
    if (Percentage >= 70):
        return 'B'
    if (Percentage >= 60):
        return 'C'
    if (Percentage >= 50):
        return 'D'
    if (Percentage >= 40):
        return 'E'
    else:
        return 'F'
    
df['Grade'] = df.apply(lambda x : GetGrade(x['Percentage'], x['All3_Pass']), axis=1)
df.Grade.value_counts()
```
