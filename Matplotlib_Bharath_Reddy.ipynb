## one style of figure I'd like to learn 

Discrete distribution as horizontal bar chart
I would be interested in learning how to make an inverted horizontal bar chart for discrete categorical data.

### This could be useful for the data to visualize
An inverted horizontal bar chart displays data where categories are represented by horizontal bars, but the bars are arranged from most frequent at the top to least frequent at the bottom. This inverts the usual order of bar charts, making it easier to quickly identify the most common categories.

### When to use it:

Discrete categorical data: When you have data that falls into distinct categories, like "Sex" with "male" and "female" or "Educ" with different educational levels.
Comparison of frequencies: When you want to compare the prevalence of different categories within your data.

Seeing the horizontal bars allows easy comparison.

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df= pd.read_csv("C:\\Users\\Bharath Reddy\\cces_example_data.csv")

df
```

```python
# Create a figure with two axes
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# title for first axis
ax1.set_title("Age Histogram")

# title second axis
ax2.set_title("Political Identity")

# empty figure
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Create a bar plot for the 'Evaluation' variable
df['Evaluation'].value_counts().sort_index()

bar_labels = ['1','2','3','4']
bar_colors = ['red','orange','lightgreen','darkgreen']

# Creating the custom-colored bars
ax.bar(bar_labels, df['Evaluation'].value_counts().sort_index(), color=bar_colors)


# Set labels and title
ax.set_xlabel('Evaluation')
ax.set_ylabel('Count')
ax.set_title('Bar Plot of Evaluation Variable')

# Show the figure
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Create a scatter plot for 'ConMediaUse' and 'ImmiPhobia'
ax.scatter(df['ConMediaUse'], df['ImmiPhobia'], alpha=0.5, color='Blue')

# Set labels and title
ax.set_xlabel('ConMediaUse')
ax.set_ylabel('ImmiPhobia')
ax.set_title('Scatter Plot of ConMediaUse vs. ImmiPhobia')

# Show the figure
plt.show()
```
