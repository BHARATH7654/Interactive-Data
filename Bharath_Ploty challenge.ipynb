```python
Water_Quality= pd.read_csv("C:\\Users\\srika\\Interactive_data\\wq.csv")
Water_Quality.head()
```

```python
# Histogram to display samples from different sites
fig = px.histogram(Water_Quality, x="Site_Id", histfunc='count', color='Site_Id', title='Count of readings from each Site Id')
fig.show()
```

```python

# Define the data and parameters for the histogram
data = Water_Quality
x = 'Year'
y = 'Dissolved Oxygen (mg/L)'
color = 'Site_Id'
barmode = 'group'
histfunc = 'avg'
text_auto = True
height = 500
title = 'Average dissolved oxygen for each site during each year'

# Create the histogram using plotly express
fig = px.histogram(data, x=x, y=y, color=color, barmode=barmode, histfunc=histfunc,
                   text_auto=text_auto, height=height, title=title)

# Display the histogram
fig.show()
```

```python

# Creating a histogram with Plotly Express
fig = px.histogram(
    Water_Quality,                                      # Dataframe containing the data
    x='Year',                                # X-axis: Year
    y='Dissolved Oxygen (mg/L)',             # Y-axis: Dissolved Oxygen
    color="Site_Id",                         # Color encoding by Site_Id
    barmode='group',                         # Grouping bars
    histfunc='avg',                          # Computing average for each bin
    facet_col='Site_Id',                     # Creating facets by Site_Id
    facet_col_wrap=3,                        # Wrapping facets in 3 columns
    text_auto=True,                          # Automatically label data points
    height=700,                              # Setting height of the plot
    title='Average Dissolved Oxygen by year for each Site'  # Setting title
)

# Displaying the plot
fig.show()
```

```python
# violin plot of water depth by site ID

# Extract unique Site IDs and corresponding data
site_ids = Water_Quality['Site_Id'].unique()
site_data = [Water_Quality[Water_Quality['Site_Id'] == site]['Water Depth (m)'].values for site in site_ids]

# Create a violin plot
plt.violinplot(site_data, showmeans=False, showmedians=True)

# Set x-axis ticks and labels
plt.xticks(ticks=np.arange(1, len(site_ids) + 1), labels=site_ids)

# Set labels and title
plt.xlabel('Site ID')
plt.ylabel('Water Depth (m)')
plt.title('Violin Plot of Water Depth by Site ID')

# Enable grid
plt.grid(True)

# Show plot
plt.show()
```

```python
# violin plot of water depth by site ID

# Create violin plot
violins = plt.violinplot(dataset=Water_Quality['Water Depth (m)'].groupby(Water_Quality['Site_Id']).apply(list), \
                         showmeans=False, showmedians=True)

# Set x-axis ticks and labels
plt.xticks(ticks=range(1, len(violins['bodies']) + 1), labels=Water_Quality['Site_Id'].unique())

# Set labels and title
plt.xlabel('Site ID')
plt.ylabel('Water Depth (m)')
plt.title('Violin Plot of Water Depth by Site ID')

# Enable grid
plt.grid(True)

# Define your own colors
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Set face color for each violin plot
for color, violin in zip(colors, violins['bodies']):
    violin.set_facecolor(color)

# Show plot
plt.show()
```

```python
# Creating a 2*2 axes with the water depth by site ID on the first column, first row

# Create a 2x2 grid of axes
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Loop through each subplot and create a violin plot
for i in range(2):
    for j in range(2):
        # Create violin plot for water depth by site ID
        site_id_data = Water_Quality['Water Depth (m)'].groupby(Water_Quality['Site_Id']).apply(list)
        violins = axs[i, j].violinplot(site_id_data, showmeans=False, showmedians=True)

        # Define your own colors
        colors = ['red', 'blue', 'green', 'orange', 'purple']

        # Set face color for each violin plot
        for color, violin in zip(colors, violins['bodies']):
            violin.set_facecolor(color)

        # Set x-axis ticks and labels
        axs[i, j].set_xticks(ticks=range(1, len(violins['bodies']) + 1))
        axs[i, j].set_xticklabels(Water_Quality['Site_Id'].unique())

        # Set labels and title
        axs[i, j].set_xlabel('Site ID')
        axs[i, j].set_ylabel('Water Depth (m)')
        axs[i, j].set_title('Violin Plot of Water Depth by Site ID')

        # Enable grid
        axs[i, j].grid(True)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()


```
